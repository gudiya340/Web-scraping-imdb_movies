```python
import numpy as np
import pandas as pd
import requests 
from bs4 import BeautifulSoup
from requests import get
from time import sleep
from random import randint
```

```python
url="https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"
```

```python
r=requests.get(url)
```

```python
r
```

```python
soup=BeautifulSoup(r.content,'html.parser')
```

```python
print(soup)
```

```python
Sno=[]
movie_name=[]
Directors=[]
Duration=[]
year=[]
ratings=[]
metascore=[]
votes=[]
Gross=[]
genre=[]
Stars=[]
certificate=[]
```

```python
movie_data=soup.findAll('div',attrs={'class':'lister-item mode-advanced'})
```

```python
for store in movie_data:
    name=store.h3.a.text
    movie_name.append(name)
    
    year_of_release=store.h3.find('span',class_='lister-item-year text-muted unbold').text.replace('(','').replace(')','')
    year.append(year_of_release)
                                  
    runtime=store.p.find('span', class_='runtime').text.replace(' min','')
    Duration.append(runtime) 
                                  
    rate=store.find('div',class_='inline-block ratings-user-rating').text.replace('\n','')
    ratings.append(rate)
    
                                  
    meta=store.find('span',class_='metascore').text if store.find('span',class_='metascore')else'^^^^'
    metascore.append(meta) 
    
    
    Directors.append(store.find('p',{"class":""}).text.split("\n")[2])
    
    nv=store.find_all('span',attrs={'name':'nv'})
    vote= nv[0].text
    votes.append(vote)
    
    gross=nv[1].text if len(nv)>1 else'_'
    Gross.append(gross) 
    
    try:
        Stars.append(store.find('p',{"class":""}).text.split("\n")[5:10])
    except:    
        Stars.append(1)
        
     
    try:
        certificate.append(store.find('span',{"class":"certificate"}).text)
        
    except:
        certificate.append(0)
    genre.append(store.find('span',{"class":"genre"}).text.replace("\n"," "))   
        
      
        
    
```

# Making a single Dataframe of imdb movies

```python
movie_data1=pd.DataFrame({'Name of movie':movie_name, 'Director of movie':Directors, 'Duration of movie':Duration, 
                       'Year of Release':year, 'Rating':ratings,'Metascore':metascore,
                       'vote':votes,'gross':Gross,'Genre':genre,'star':Stars,'Certificate':certificate})
```
